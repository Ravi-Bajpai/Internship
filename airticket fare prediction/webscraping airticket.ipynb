```python
#function has been define,so that it can be called for different source and destination
def flightticket():
    flightname = driver.find_elements(By.XPATH,'//span[@class="txt-r4 ng-binding"]')
    flight_name=[]
    for i in flightname:
        flight_name.append(i.text)
    airportname_tag = driver.find_elements(By.XPATH,'//span[@class="txt-r3-n ng-binding"]')    
    airportname=[]
    for i in airportname_tag:
        airportname.append(i.text)
    source=[]
    source=airportname[0::2]  
    destination=[]
    destination=airportname[1::2]
    fl_tag = driver.find_elements(By.XPATH,'//span[@class="dura_md ng-binding"]')
    fl=[]
    for i in fl_tag:
        fl.append(i.text)
    st_tag = driver.find_elements(By.XPATH,'//span[@class="dura_md2 ng-scope"]')
    stop=[]
    for i in st_tag:
        stop.append(i.text)
    price_tag = driver.find_elements(By.XPATH,'//div[@class="col-md-2 col-sm-2 col-xs-5 mr5 cle"]')    
    price=[]
    for i in price_tag:
        price.append(i.text.split('\n')[0])
    fltime_tag=driver.find_elements(By.XPATH,'//span[@class="txt-r2-n ng-binding"]')    
    fltime=[]
    for i in fltime_tag:
        fltime.append(i.text)
    departure=[]
    departure=fltime[0::2] 
    arrival=[]
    arrival=fltime[1::2]
    df1=pd.DataFrame()
    df1['FLIGHT NAME']=flight_name
    df1['SOURCE']=source
    df1['DESTINATION']=destination
    df1['ARRIVAL']=arrival
    df1['DEPARTURE']=departure
    df1['TIME OF TRAVEL']=fl
    df1['NO. OF STOP']=stop
    df1['PRICE']=price
    return(df1)    
```

```python
df1=flightticket()
df1["DATE"]=date
dffinal=pd.DataFrame()
dffinal=dffinal.append(df1,ignore_index = True)
```

```python
def makemytrip():
    allinfo = driver.find_elements(By.XPATH,'//div[@class="makeFlex simpleow"]')
    airlinename=[]
    departure=[]
    source=[]
    timeofjourney=[]
    stop=[]
    arrival=[]
    destination=[]
    price=[]
    allinform=[]
    for i in allinfo:
        allinform.append(i.text)
        airlinename.append(i.text.split('\n')[0])
        departure.append(i.text.split('\n')[2])
        source.append(i.text.split('\n')[3])
        timeofjourney.append(i.text.split('\n')[4])
        stop.append(i.text.split('\n')[5])
        arrival.append(i.text.split('\n')[6])
        destination.append(i.text.split('\n')[7])
        price.append(i.text.split('\n')[8])
    df2=pd.DataFrame()
    df2['FLIGHT NAME']=airlinename
    df2['SOURCE']=source
    df2['DESTINATION']=destination
    df2['ARRIVAL']=arrival
    df2['DEPARTURE']=departure
    df2['TIME OF TRAVEL']=timeofjourney
    df2['NO. OF STOP']=stop
    df2['PRICE']=price                  
    return(df2)
```

```python
driver=webdriver.Chrome(r'D:\flip robo\chromedriver_win32\chromedriver.exe')
```

```python
url="https://www.cleartrip.com/flights"
driver.get(url)
```

```python
from selenium.webdriver.common.by import By
```

```python
def cleartrip():
    flight1=driver.find_elements(By.XPATH,'//div[@data-testid = "airlineBlock"]')
    airlinename=[]
    departure=[]
    source=[]
    timeofjourney=[]
    stop=[]
    arrival=[]
    destination=[]
    price=[]
    flight_name1=[]
    for i in flight1:
        flight_name1.append(i.text)
        airlinename.append(i.text.split('\n')[0])
        departure.append(i.text.split('\n')[3])
        
        timeofjourney.append(i.text.split('\n')[4])
        stop.append(i.text.split('\n')[5])
        arrival.append(i.text.split('\n')[6])
        
    price_tag=driver.find_elements(By.XPATH,'//p[@class="m-0 fs-5 fw-700 c-neutral-900 false"]')   
    for i in price_tag:
        price.append(i.text)
    df7=pd.DataFrame()
    df7['FLIGHT NAME']=airlinename
    
    df7['DEPARTURE']=departure
    df7['ARRIVAL']=arrival
    
    df7['TIME OF TRAVEL']=timeofjourney
    df7['NO. OF STOP']=stop
    df7['PRICE']=price    
    return(df7)
```
